```python
import pandas as pd
```

```python
df = pd.read_csv("../data/wages_hours.csv", delimiter="\t")
df.head()
```

```python
df.isnull().values.any()
```

```python
import numpy as np
df_fixed = df.replace("*", np.nan)
df_fixed.head()
```

```python
df_missing = df_fixed[df_fixed.isnull().any(axis=1)]
df_missing.head(100)
```

```python
import seaborn as sns
```

```python
sns.histplot(df_fixed, x="ERNO")
```

```python
df_fixed["ERNO"] = df_fixed["ERNO"].astype("float")

df_fixed.head(100)
```

```python
df_median_imputed = df_fixed.copy()
df_median_imputed["ERNO"] = df_median_imputed["ERNO"].fillna(df_median_imputed["ERNO"].median())
df_median_imputed.head(100)
```
